```python
import imutils
```

```python
## Cascade Classifier algorithm in OpenCV for face detect
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

## Cascade Classifier algorithm in OpenCV for smile detect
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')

## Cascade Classifier algorithm in OpenCV for eye detect
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
## read the img in array
img = cv2.imread('face.jpg')

print(img)
## Resize the image
img = imutils.resize(img,width=1200)

## Converting BGR img to Gray space
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray color',gray)
```

```python
plt.imshow(gray)
```

```python
## use face cascade to extract the face cordinates
# 1.3 --> Threshold, 5--> no of neighbouring pixels
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

# print no of faces
print(len(faces))
```

```python
# print the vlues
print(faces) # (x,y,w,h)
```

```python
## read the img in array
img = cv2.imread('faces.jpg')

print(img)
## Resize the image
img = imutils.resize(img,width=1200)

## Converting BGR img to Gray space
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
%matplotlib auto
cv2.imshow('gray color',gray)
```

```python
plt.imshow(gray)
```

```python
## use face cascade to extract the face cordinates
# 1.3 --> Threshold, 5--> no of neighbouring pixels
faces = face_cascade.detectMultiScale(gray, 1.05, 3)

# print no of faces
print(len(faces))
```

```python
# print the vlues
print(faces) # (x,y,w,h)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (225,225,225),1)
    roi_gray = gray [y:y+h, x:x+w]
    roi_color = img [y:y+h, x:x+w]
```

```python
plt.imshow(img)
```

```python
%matplotlib auto
cv2.imshow('predict faces',img)
```
